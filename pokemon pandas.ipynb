```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df2 = pd.read_csv('C:/Users/livia babu/Downloads/car dataset.csv')

# Count the number of sales by production year
sales_counts = df2['Prod. year'].value_counts()

# Get the top 5 production years with the highest sales
top_5_years = sales_counts.nlargest(5)

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_5_years.values, labels=top_5_years.index, autopct='%1.1f%%')
plt.title('Top 5 Production Years with the Highest Sales')

# Save the chart


# Get the top 5 years
top_5_years.index
```

```python
import matplotlib.pyplot as plt

# Calculate the count of cars by color
color_counts = df2['Color'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(color_counts.index, color_counts.values, color='skyblue')
plt.xlabel('Color')
plt.ylabel('Count')
plt.title('Count of Cars by Color')
plt.xticks(rotation=90)


# Find the color with the most cars
max_color = color_counts.idxmax()
max_color
```
